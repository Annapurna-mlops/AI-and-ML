```python
# Some time we may get output like  
# Hello 
# Hi 
# HelloHi
# HiHello
# this is because thread1 and thread 2 are sending request to cup at same time and CPU is processing at same time.
# This is called collision. To avoid these collision we can use sleep method to schedule a thread to sleep for sometime


from threading import Thread
from time import sleep


class Hello(Thread):
    def run(self):
        for i in range(500):
            print("Hello")

            sleep(1)  # sleep for 1 sec

class Hi(Thread):
    def run(self):
        for i in range(500):
            print("Hi")
            sleep(1)  # sleep for 1 sec

t1 = Hello()
t2 = Hi()
# if we use run method, internally it's not creating two threads
# To create a new thread from Main thread, we have to use Start method
# Start method internally will call run method. You can check run method in Thread class
t1.start() 
sleep(0.2)  # To avoid thread collision. 
t2.start()

print("How are you")   # sometimes you will get this output in the middle of t1 and t2 threads output 
# We have to ask main method to wait until t1 and t2 thread joins 

t3 = Hello()
t4 = Hi()
t3.join()
t4.join()

print("checking Main thread ") 
```

```python
# thread synchronization within objects

"""
wait(): The calling thread releases the lock on the Condition object and waits until another thread calls notify() or notifyAll() on the same Condition object. It reacquires the lock before continuing.
notify(): Wakes up a single waiting thread on the Condition object. If no threads are waiting, the call has no effect.
notifyAll(): Wakes up all threads waiting on the Condition object.

Use Cases:
Complex synchronization scenarios where multiple threads need to coordinate access to shared resources.

Important Note: These methods require careful use to avoid race conditions and deadlocks.

Using a lock (or more specifically, a threading.Lock object) is mandatory when working with wait() and notify()/notifyAll() methods in Python for thread synchronization.

The lock ensures that only one thread can access the shared resource (the Condition object itself and potentially other data it protects) at a time. This prevents race conditions where multiple threads might modify the data in unexpected ways.
"""

import threading

class Buffer:
  def __init__(self, max_size):
    self.lock = threading.Lock()
    self.condition = threading.Condition(self.lock)
    self.items = []
    self.max_size = max_size

  def produce(self, item):
    with self.lock:
      while len(self.items) >= self.max_size:
        self.condition.wait()
      self.items.append(item)
      self.condition.notify()  # Wake up a waiting consumer

  def consume(self):
    with self.lock:
      while len(self.items) == 0:
        self.condition.wait()
      item = self.items.pop(0)
      self.condition.notify()  # Wake up a waiting producer
      return item

# Producer and consumer threads (implementation omitted for brevity)

```
