````python
# Class and Method
class Person:
  # Attribute -- can be Static 
  def __init__(self, name, age):
    self.name = name   # define or intial new attibutes 
    self.age = age

  def introduce(self):
    """Introduces the person."""
    print(f"Hi, I'm {self.name} and I am {self.age} years old.")
    #Methods can access and manipulate the data (attributes) of the object they are called on.
person1 = Person("Bob", 30)
person1.introduce()  # Output: Hi, I'm Bob and I am 30 years old.
````

Anonymous Functions (Lambda Functions)
------------------------
- **Definition:** Use the lambda keyword followed by arguments (comma-separated), a colon (:), and a single expression that defines the function's return value.

````python
# Lambda function to calculate the square of a number
def square(x):
    x = x* x
    return x


square1 = lambda x: x * x

result = square1(4)
print(result)  # Output: 16
````

**Key Differences:**

- **Names:** Regular functions have names, while lambda functions are anonymous (nameless).
- **Complexity:** Regular functions can have complex logic with multiple statements, while lambda functions are limited to a single expression.

**Use Cases:**

- **Regular Functions:** Suitable for defining reusable, named blocks of code that might have multiple statements or return values.
- **Lambda Functions:** Ideal for short, simple operations where a named function would be cumbersome. They are often used as arguments to higher-order functions (functions that take other functions as arguments) like `map`, `filter`, and `reduce`.

```python
# Using lambda with map
numbers = [1, 2, 3, 4, 5]
squares = list(map(lambda x: x * x, numbers))
print(squares)  # Output: [1, 4, 9, 16, 25]
```

**In Summary:**

- Regular functions provide a structured way to create reusable code blocks with names, arguments, and potentially complex logic.
- Lambda functions offer a concise syntax for short, anonymous operations, often used with higher-order functions.

In Python, `*args` and `**kwargs` are special operators used within function definitions to handle variable-length arguments. They provide flexibility in how many and what kind of arguments you can pass to a function.

**1. `*args` (Arbitrary Arguments):**

- **Purpose:** Used to capture an arbitrary number of positional arguments (arguments passed in the order they appear in the function call) into a tuple within the function.
- **Syntax:**
   - In function definition: `def my_function(*args):`
   - In function call: `my_function(arg1, arg2, ..., argN)`
- **Example:**

```python
def print_all(*args):
  """Prints all arguments passed to the function."""
  for arg in args:
    print(arg)

print_all("Hello", 10, True)  # Output: Hello, 10, True
```

**2. `**kwargs` (Keyword Arguments):**

- **Purpose:** Used to capture an arbitrary number of keyword arguments (arguments passed with a key-value pair syntax) into a dictionary within the function.
- **Syntax:**
   - In function definition: `def my_function(**kwargs):`
   - In function call: `my_function(key1=value1, key2=value2, ..., keyN=valueN)`
- **Example:**

```python
def greet(name, **options):
  """Greets someone with optional message and title."""
  greeting = f"Hello, {name}"
  if "message" in options:
    greeting += f" ({options['message']})"
  if "title" in options:
    greeting = f"{options['title']} {greeting}"
  print(greeting)

greet("Alice", message="How are you?", title="Ms.")  # Output: Ms. Hello, Alice (How are you?)
```

**Combining `*args` and `**kwargs`:**

- You can use both `*args` and `**kwargs` in the same function definition, but `*args` must come before `**kwargs`.
- This allows you to capture a mix of positional and keyword arguments.

**Common Use Cases:**

- **Handling Unknown Numbers of Arguments:** Useful for functions that can take a varying number of arguments, like printing elements of a list or performing calculations on multiple values.
- **Customizable Functions:** Enable functions to accept additional options or configurations through keyword arguments.
- **Unpacking Collections:** When used within function calls (not definitions), `*args` and `**kwargs` can unpack collections like tuples or dictionaries as separate arguments.

**In Summary:**

`*args` and `**kwargs` provide powerful mechanisms for creating flexible functions that can handle various argument scenarios in Python. By understanding these operators, you can write more adaptable and reusable code.
